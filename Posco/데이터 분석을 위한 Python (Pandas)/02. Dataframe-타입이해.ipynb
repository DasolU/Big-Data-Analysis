        - Ìï¥Îãπ Î≤îÏ£ºÏóê Ìï¥ÎãπÌïòÎ©¥ 1, ÏïÑÎãàÎ©¥ 0ÏúºÎ°ú Ï±ÑÏö∞Îäî Ïù∏ÏΩîÎî© Í∏∞Î≤ï

            - drop_first: Î∂àÌïÑÏöîÌïú Î≥ÄÏàòÎ•º Ï§ÑÏó¨Ï§ÄÎã§

```python
# Ïà´ÏûêÌòï Îç∞Ïù¥ÌÑ∞Î•º Î≤îÏ£ºÌòïÏúºÎ°ú Î≥ÄÌôò
import math

def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10  

df = pd.read_csv(direc)
df['Age'].apply(age_categorize)
```

```python
# Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨
# one-hot incoding
pd.get_dummies(df)
```

```python
# PclassÍ∞Ä Ïù¥Ï†ÑÏóêÎäî 1,2,3 Ï§ëÏùò ÌïòÎÇòÏùò Í∞íÏùÑ Í∞ÄÏ°åÎã§.
# get_dummiesÏóêÏÑú columnsÎ•º ÏßÄÏ†ï=> 1,2,3 Í∞Å Í∞íÏùÑ columnÏúºÎ°ú Î≥ÄÌôòÌïòÏó¨ 0 ÎòêÎäî 1Î°ú Ï±ÑÏõåÏ§ÄÎã§.
pd.get_dummies(df, columns = ['Pclass', 'Sex'])
```

```python
# drop_first=True
# Î∂àÌïÑÏöîÌïú Î≥ÄÏàòÎ•º Ï§ÑÏó¨Ï§ÄÎã§.
# 1,2,3 Ï§ëÏóê 2ÏôÄ 3Ïù¥ 0Ïù∏ Í≤ΩÏö∞ 1Ïóê Ìï¥ÎãπÌïòÎØÄÎ°ú Î≥ÄÏàòÎ•º Ï§ÑÏó¨Ï§ÄÎã§.
pd.get_dummies(df, columns = ['Pclass', 'Sex'], drop_first=True)
```
