# 1-2. 프로그래밍 과목의 중간고사와 기말고사 점수를 입 력 받아 평균과 학점을 구하는 프로그램

```python
def even_or_odd(mid, final):
    mean = (mid+final)/2
    if mean>=90:
        grade = 'A'
    elif mean>=80:
        grade = 'B'
    elif mean>=70:
        grade = 'C'
    elif mean>=60:
        grade = 'D'
    else:
        grade = 'F'
    
    print('Average: ', mean)
    print('Grade: ', grade)

mid = int(input("Enter your midterm score: "))
final = int(input("Enter your final score: "))

even_or_odd(mid, final)
```

# 1-3. 합격여부 판단 프로그램


```python
def is_pass(eng, math):
    total = eng+ math

    if total>=110:
        if eng<40:
            print('불합격: 영어 점수 부족')
            return
        if math<40:
            print('불합격: 수학 점수 부족')
            return
        else:
            print('합격')
    else: 
        print('불합격: 총합 점수 부족')   

for _ in range(5):
    eng = int(input("영어 점수 입력: "))
    math = int(input("수학 점수 입력: "))
    is_pass(eng, math)
```

# 1-4. 세 개의 정수를 입력 받아, 가장 큰 수만 출력하는 프로그 램

```python
def find_max(x1, x2, x3):
    max_num = 0
    if x1>=x2:
        max_num = x1
    else:
        max_num = x2
    if max_num > x3:
        print('가장 큰 수는 %d입니다' %max_num)
    else:
        print('가장 큰 수는 %d입니다' %x3)
        
x1, x2, x3 = map(int,input("세 개의 수를 입력하시오: ").split())    
find_max(x1, x2, x3)
```

# 2-1-1. 문자열 출력: while문 이용

```python
def print_word(word):
    ct=0
    while ct<=len(word)-1:
        print(word[ct])
        ct+=1
word = 'Python'
print_word(word)
```

# 2-1-2. 문자열 출력: for문 이용

```python
def print_word(word):
    for i in word:
        print(i)
word = 'Python'
print_word(word)
```

# 2-2. 출력 연습
- print(sep = ' ', end=',')

```python
def print_commma(line):
    a = ', '
    print(a.join(list(map(str,list(range(1,11)[::-1])))), end=", ")
    print(line)

print_commma('Hello new year!!')
```

# 2-3. 양의 두 정수 a,b를 입력 받아,a부터 b까지의 정 수의 합을 구하여 출력하는 프로그램

```python
def sum_from_atob(a,b):
    total = 0
    for i in range(a,b+1):
        total+= i    
    print('The sum from {} to {} is {}'.format(a,b,total))

a, b = map(int,input("Enter two integers: ").split())    
sum_from_atob(a,b)
```

# 2-4. 주어진 문자열에 문자a가 몇 개 있는지 구하는 프 로그램 (for문 사용)

```python
def find_a(word):
    ct = 0
    for i in word:
        if i =='a':
            ct+=1
    print('문자열 {}에 문자a가 {}에 있습니다.'.format(word, ct))
    
word = 'banana'
find_a(word)
```

# 2-8. 구구단 출력1

# 2-9. 구구단 출력2 (중첩 반복문)

```python
def multiplication_table_all():
    for i in range(2,10):
        print('=={}단=='.format(i))
        for j in range(1,10):
            print('{} * {} = {}'.format(i,j,i*j))

multiplication_table_all()
```

# 2-10. 구구단 출력2 (중첩 반복문)

```python
def multiplication_table_all_collumnwise():
    for i in range(1,10):
        for j in range(2,10):
            print('%d * %d = %2d' %(j, i, i*j), end=' ')
        print('\n')

multiplication_table_all_collumnwise()
```

```python
def mean_max_min():
    flag = True
    save = []
    while flag:
        is_done = input('Enter a number: ') 
        if is_done =='done':
            flag = False
        else:
            save.append(float(is_done))
    print(save)
    print('Average: ', sum(save)/len(save))
    print('Maximum: ', max(save))
    print('Minimum: ', min(save))

mean_max_min()
```
