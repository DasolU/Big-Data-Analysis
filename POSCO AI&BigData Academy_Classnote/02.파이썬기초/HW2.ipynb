# 1-2. í”„ë¡œê·¸ëž˜ë° ê³¼ëª©ì˜ ì¤‘ê°„ê³ ì‚¬ì™€ ê¸°ë§ê³ ì‚¬ ì ìˆ˜ë¥¼ ìž… ë ¥ ë°›ì•„ í‰ê· ê³¼ í•™ì ì„ êµ¬í•˜ëŠ” í”„ë¡œê·¸ëž¨

```python
def even_or_odd(mid, final):
    mean = (mid+final)/2
    if mean>=90:
        grade = 'A'
    elif mean>=80:
        grade = 'B'
    elif mean>=70:
        grade = 'C'
    elif mean>=60:
        grade = 'D'
    else:
        grade = 'F'
    
    print('Average: ', mean)
    print('Grade: ', grade)

mid = int(input("Enter your midterm score: "))
final = int(input("Enter your final score: "))

even_or_odd(mid, final)
```

# 2-3. ì–‘ì˜ ë‘ ì •ìˆ˜ a,bë¥¼ ìž…ë ¥ ë°›ì•„,aë¶€í„° bê¹Œì§€ì˜ ì • ìˆ˜ì˜ í•©ì„ êµ¬í•˜ì—¬ ì¶œë ¥í•˜ëŠ” í”„ë¡œê·¸ëž¨

```python
def sum_from_atob(a,b):
    total = 0
    for i in range(a,b+1):
        total+= i    
    print('The sum from {} to {} is {}'.format(a,b,total))

a, b = map(int,input("Enter two integers: ").split())    
sum_from_atob(a,b)
```

```python
def mean_max_min():
    flag = True
    save = []
    while flag:
        is_done = input('Enter a number: ') 
        if is_done =='done':
            flag = False
        else:
            save.append(float(is_done))
    print(save)
    print('Average: ', sum(save)/len(save))
    print('Maximum: ', max(save))
    print('Minimum: ', min(save))

mean_max_min()
```
