# 1. set

```python
class makeset:
    def __init__(self, inlist=[]):
    
        if len(inlist)==0:
            self.set = []

        else:
            myset = []
            q = inlist
            while q:
                n = q.pop(0) 
                if n not in myset:
                    myset.append(n)

            self.set = myset 

    def add(self, elem):
        if elem not in self.set:
            self.set.append(elem)

    def discard(self, elem):
        if elem in self.set:
            self.set.remove(elem) #시간 복잡도가 O(n)인 list연산, 어떻게 효율화 할 수 있을지 고민해보자.
        
    def clear(self):
        self.set = []

    def __len__(self):
        self.count = len(self.set)
        return self.count
    
    def __str__(self):
        sep = ', '
        return '{'+sep.join(list(map(str,self.set)))+'}'
    
    def __contains__(self, elem):
        if elem in self.set:
            return True
        else:
            return False
        
    def __le__(self, other):
        flag = True
        for ele in self.set:
            if ele not in other.set:
                flag = False
        return flag
    
    def __ge__(self, other):
        flag = True
        for ele in other.set:
            if ele not in self.set:
                flag = False
        return flag

    def __or__(self, other):
#         copy = self.setcopy
        copy = [i for i in self.set]
        for ele in other.set:
            if ele not in copy:
                copy.append(ele)
        return makeset(copy)

    def __and__(self, other):
        join = []
        for ele in self.set:
            if ele in other.set:
                join.append(ele)
        return makeset(join)
    
    def __sub__(self, other):
        copy = [i for i in self.set]
        for ele in other.set:
            if ele in copy:
                copy.remove(ele)
        return makeset(copy)    

    
#     |=: 자기 자신에 합집합
    def __ior__(self, other):
        for ele in other.set:
            if ele not in self.set:
                self.set.append(ele)
        return self
    
    # &=: 자기 자신에 교집합
    def __iand__(self, other):
        dummy = []
        for ele in self.set:
            if ele not in other.set:
                dummy.append(ele)
        for i in dummy:
            self.set.remove(i)
                
        return self
    
    # -=: 자기 자신에 차집합
    def __isub__(self, other):
        dummy = []
        for ele in other.set:
            if ele in self.set:
                dummy.append(ele)

        for i in dummy:
            self.set.remove(i)
        return self
```

```python
a = makeset([1, 2, 3, 4])
b = makeset([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()


print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()
```

```python
a = makeset([1, 2, 3])
b = makeset([3, 4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = makeset([1, 2, 3])
b = makeset([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = makeset([1, 2, 3])
b = makeset([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
