# 1-3. ì›ì˜ ë‘˜ë ˆì™€ ë„“ì´ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨

```python
def cal_circle(radius):
    pi = 3.141592
    circle_length = 2*pi*radius
    width = pi*radius**2
    print('ì› ë‘˜ë ˆ: {:.2f}'.format(circle_length))
    print('ì› ë„“ì´: {:.2f}'.format(width))

radius = int(input("ë°˜ì§€ë¦„ì„ ì…ë ¥í•˜ì‹œì˜¤: "))
cal_circle(radius)
    
```

# 1-4. ì‚¬ì¹™ì—°ì‚° ë° ë‚˜ë¨¸ì§€ ì—°ì‚°



```python
def cal_integer(one, two):
    print('{0} + {1} = {2}'.format(one, two, one+two))
    print('{0} - {1} = {2}'.format(one, two, one-two))
    print('{0} * {1} = {2}'.format(one, two, one*two))
    print('{0} / {1} = {2}'.format(one, two, one/two))
    print('{0} % {1} = {2}'.format(one, two, one%two))

 
one, two = map(int,input("Enter two integers: ").split())    
cal_integer(one, two)
```

# 1-5. ì‚¬ì¹™ì—°ì‚° ë° ë‚˜ë¨¸ì§€ ì—°ì‚°


```python
def sum_and_mean(one, two):
    print('The sum of {0} and {1} is {2}'.format(one, two, one+two))
    print('The average of {0} and {1} is {2}'.format(one, two, (one+two)/2))

one, two = map(int,input("Enter two integers: ").split())    
sum_and_mean(one, two)
```

```python
def print_date(inputstr):
    year, month, day = map(int,inputstr.split('/'))
    after = 10
    print('ì…ë ¥í•œ ë‚ ì§œì˜ {0}ë…„ í›„ëŠ” {1}ë…„ {2}ì›” {3}ì¼'.format(after, year+after, month, day)) 

date = input("ë‚ ì§œ(ì—°/ì›”/ì¼)ì…ë ¥: ")
print_date(date)
```
