## 2. 실습 (기술통계): 자동차 연비 데이터
### 해석
corolla는 연비가 높음에도 불구하고 전체 개수에서 차지하는 정도가 약 2.13%으로 낮은 비중을 차지하는 것으로 보입니다.

```python
df = ds_mycars[['model','mpg','highway_mileage']]
# count
count = df['model'].value_counts().sort_index()
# cumulative counts
cumcnt = np.cumsum(count)
# percentage
percent = count/sum(count) *100
# cumulative percentage
cumpct = np.cumsum(percent)

count_data = pd.DataFrame({'count': count, 'percent': percent,'cumcnt':cumcnt,  'cumpct':cumpct})
count_data
```
