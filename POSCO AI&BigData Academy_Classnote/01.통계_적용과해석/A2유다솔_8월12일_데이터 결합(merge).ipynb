# 1. 연습 데이터 생성
fitness_1, fitness_2 파일을 생성하기 위해 

fitness_prac.csv 파일을 사용하였습니다.

## fitness_prac.csv 파일을 불러오기

```python
import pandas as pd

dir_ ='/Users/dasol/Documents/GitHub/Big-Data-Analysis/POSCO_classnote/'
df = pd.read_csv(dir_+'02.BigData분석/dataHandling-main/data/data/fitness_prac_n.csv', engine='python')
df
```

## 이상치 처리
gender==5 => 쿼리를 사용하여 라벨이 5이하인 데이터만 반환하여 사용


## gender 라벨이 숫자이므로 여성/남성으로 변환


gender==2 => 여성으로 변환

gender==1 => 남성으로 변환

```python
df = df.query("gender <5")
df.replace({'gender': 2}, '여성', inplace=True)
df.replace({'gender': 1}, '남성', inplace=True)
df.head()
```

## FITNESS_1와 FITNESS_2 데이터 프레임 생성

```python
left = df[['name', 'gender', 'age', 'agegroup', 'weight']]

right = df[['name', 'gender', 'age', 'oxy', 'runtime',
       'runpulse', 'restpulse', 'maxpulse']]
```

# 2. merge를 이용한 결합
병합할때 사용할 index를 'name'으로 사용하여 병합하였습니다.

```python
result = pd.merge(left, right, on=['name'])
result
```
