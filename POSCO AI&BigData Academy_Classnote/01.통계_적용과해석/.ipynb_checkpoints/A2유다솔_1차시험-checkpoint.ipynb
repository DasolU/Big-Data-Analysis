## 21. 정규분포의 확률 계산
### 해석
해석: 'A 부서', 'B 부서', 'C 부서', 'D 부서'의 p-value가 모두 >0.05 => 'A 부서', 'B 부서', 'C 부서', 'D 부서'에 대해서는 정규 분포를 따르고 있다.

```python
### 정규성 검정
df = pd.read_csv('./1차시험_주관식 데이터/전표처리시간1', engine="python")
df.head()
print(df.mean())
```

```python
### 정규성 검정
feature = ['A 부서', 'B 부서', 'C 부서', 'D 부서']
for f in feature:
    data= df[f]
    statistic, p = stats.shapiro(data)
    print("{}의 Shapiro-wilk test: statistic={}, p-value={}".format(f,statistic, p))
    # 해석: p-value>0.05 => Supp1에 대해서는 정규 분포를 따르고 있다.
```

```python
datas = []
feature = ['A 부서', 'B 부서', 'C 부서', 'D 부서']

for f in feature:
    data= df[f].values
    datas.extend(data)
# print(datas)

mu = np.mean(datas)
sigma = np.std(datas , ddof=1)
x = 35

# 누적 확률 값: CDF
prob=stats.norm.cdf(x,mu,sigma)
print('x({})까지 누적 확률 값: '.format(x), prob.round(3))

# 35이상의 누적확률값 계산
print('{0}이상 %범위: {1:.1f}%'.format(x, (1-prob)*100))
```

## 21. 정규분포의 확률 계산
### 해석
    - 180cm인 남성의 키는 상위 15.9%에 해당하며, 남성 100명 중 약 16명 정도가 180cm이상의 키를 가진다고 해석된다.

```python
mu = 175
sigma = 5
x = 180

# x(180)까지 누적 확률 값: CDF
prob=stats.norm.cdf(x,mu,sigma)
print('x({})까지 누적 확률 값: '.format(x), prob.round(3))

# 180이상의 누적확률값 계산
print('{0}cm인 남성의 %범위: {1:.1f}%'.format(x, (1-prob)*100))
```

#### 3-1.잔차의 **정규성** 검정:

```python
df = pd.read_csv('./1차시험_주관식 데이터/단순선형회귀.csv', engine="python")
df.head()
```

```python
reg_model = smf.ols('Y ~ X', data=df) # ordinary least squares (y ~ x, data=)
reg_result = reg_model.fit()
print(reg_result.summary())
```

### 3. 잔차의 기본 가정 확인
#### 3-2.잔차의 **등분산성** 검정

해석:
잔차들이 0을 중심으로 각각 **y = 0을 기준으로 랜덤하게 분포**하고 모두 +2, -2 범위에 포함된다.=>등분산성 만족

1. 등분산성/독립성 검정 결과, 각각 y = 0을 기준으로 랜덤하게 분포하여 가정을 만족한다.
2. 정규성검정 결과, 정규분포 직선을 중심으로 데이터가 분포하고 검정 통계량이 귀무가설을 채택하여 정규분포를 따른다고 할 수 있다.

```python
plt.scatter(reg_result.fittedvalues, reg_result.resid)
plt.xlabel('예측값')
plt.ylabel('residual')
plt.axhline(y=0, linestyle = '--', c='r')
plt.show()
plt.close()
```
