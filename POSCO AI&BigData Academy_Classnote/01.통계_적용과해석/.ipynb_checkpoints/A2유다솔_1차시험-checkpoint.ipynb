## 21.

: 'A ', 'B ', 'C ', 'D ' p-value  >0.05 => 'A ', 'B ', 'C ', 'D '     .

```python
###  
df = pd.read_csv('./1_ /1', engine="python")
df.head()
print(df.mean())
```

```python
###  
feature = ['A ', 'B ', 'C ', 'D ']
for f in feature:
    data= df[f]
    statistic, p = stats.shapiro(data)
    print("{} Shapiro-wilk test: statistic={}, p-value={}".format(f,statistic, p))
    # : p-value>0.05 => Supp1     .
```

```python
datas = []
feature = ['A ', 'B ', 'C ', 'D ']

for f in feature:
    data= df[f].values
    datas.extend(data)
# print(datas)

mu = np.mean(datas)
sigma = np.std(datas , ddof=1)
x = 35

#   : CDF
prob=stats.norm.cdf(x,mu,sigma)
print('x({})   : '.format(x), prob.round(3))

# 35  
print('{0} %: {1:.1f}%'.format(x, (1-prob)*100))
```
