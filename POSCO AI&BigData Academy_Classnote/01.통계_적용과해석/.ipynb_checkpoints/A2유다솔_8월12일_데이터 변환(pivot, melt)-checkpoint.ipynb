# 운동에 따른 혈당 산소 요구량 데이터
## gender와 agegroup별 weight의 평균 도출

# 1. 연습 데이터 생성 

## fitness_prac.csv 파일을 불러오기

```python
import pandas as pd
dir_ ='/Users/dasol/Documents/GitHub/Big-Data-Analysis/POSCO_classnote/'
df = pd.read_csv(dir_+'02.BigData분석/dataHandling-main/data/data/fitness_prac_n.csv', engine='python')
df.head()
```

## 이상치 처리
gender==5 => 쿼리를 사용하여 라벨이 5이하인 데이터만 반환하여 사용

총 데이터 개수 30개

```python
df = df.query("gender <5")
df.replace({'gender': 2}, '여성', inplace=True)
df.replace({'gender': 1}, '남성', inplace=True)
df
```

# 2. 성별과 나이대에 따른 weight 평균

```python
df_test = df[['agegroup', 'gender', 'weight']]
df_test.pivot_table(df_test, 'gender', 'agegroup')
```

# 3. pivot을 이용한 변환 

### pivot: index, column, values를 지정하여 pivot

```python
df_test.pivot_table(df_test,'agegroup', 'gender')
```
