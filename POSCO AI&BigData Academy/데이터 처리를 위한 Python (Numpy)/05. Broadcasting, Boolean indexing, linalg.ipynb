# Boolean indexing

```python
import numpy as np
x = np.random.randint(1,10, size=5)
print(x)
```

```python
even_mask = x % 2 ==0
print(even_mask)
print(x[even_mask])
print(x[x % 2 ==0])
```

```python
x[(x<3) | (x%2==0)]
```

# linalg

```python
x = np.random.rand(3,3)
print(x)
print(x@np.linalg.inv(x))
```

```python
a = np.array([[1,1],[2,4]])
b = np.array([25,64])

x = np.linalg.solve(a, b)
print(x)
np.allclose(a@x, b)
```
