## method type
- ê°ì²´ë¡œ í˜¸ì¶œ: instance method
- class í˜¸ì¶œ: static method/class method

```python
# ê°ì²´ë¡œ í˜¸ì¶œ: instance method
class counter():
    def __init__(self):
        self.num = 0
    def increment(self): # method: ì†ì„±ì„ ë³€ê²½í•œë‹¤.
        self.num +=1
    def print_current_value(self): # í•­ìƒ selfë¥¼ ê°€ì§€ê³  ìˆë‹¤.
        print('í˜„ì¬ ê°’ì€ ', self.num)
        
c1 = counter()
c1.increment()
c1.print_current_value()
c1.increment()
c1.print_current_value()

c2 = counter() # c2ëŠ” ë…ë¦½ì ìœ¼ë¡œ ê°’ì„ ìœ ì§€í•˜ê³  ìˆë‹¤.
c2.print_current_value()
```

```python
# class í˜¸ì¶œ: static method
class Math: 
    @staticmethod # python ë°ì½”ë ˆì´í„°
    def add(a, b): # selfê°€ ì—†ë‹¤.
        return a+b
    @staticmethod
    def multiply(a,b):
        return a*b

# ê°ì²´ë¥¼ ìƒì„±í•˜ì§€ ì•Šê³ 
#m = Math() 
#m.add(10,20)
# class ì´ë¦„ìœ¼ë¡œ ë¶€ë¥¼ ìˆ˜ ìˆë‹¤.
Math.add(10,20)
Math.multiply(10,20)
```

    - ìƒˆë¡œìš´ classì— ë©”ì†Œë“œê°€ ì—†ë”ë¼ë„ ë¶€ëª¨ classì— ì •ì˜í•´ë‘ë©´ ê·¸ë˜ë„ ì´ìš©í•  ìˆ˜ ìˆë‹¤.
    - method overwriting: 
        - ë¶€ëª¨ classì—ì„œ ìƒì†í•œ methodë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  ì¬ì •ì˜í•´ì„œ ì‚¬ìš©.
        - ì¬ì •ì˜ í•˜ëŠ” ìˆœê°„ ë¶€ëª¨ classì—ì„œ ì¤‘ë³µë˜ëŠ” methodë¥¼ ì‚¬ìš©í•  ìˆ˜ ì—†ê²Œ ëœë‹¤.
            - super().method():ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¶€ëª¨ class ê¸°ëŠ¥ë„ ì‚¬ìš© ê°€ëŠ¥


```python
# Student, Employeeì— ë™ì¼í•˜ê²Œ ì‚¬ìš©ë˜ëŠ” ë©”ì†Œë“œë¥¼ Person í´ë˜ìŠ¤ë¡œ ë¹¼ì¤€ë‹¤.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def eat(self, food):
        print('{}ì€ {}ë¥¼ ë¨¹ëŠ”ë‹¤.'.format(self.name, food))
    def work(self, minute):
        print('{}ì€ {}ë¶„ë™ì•ˆ ì¼í•œë‹¤.'.format(self.name, minute))

class Student(Person): # eat methodê°€ ì—†ì§€ë§Œ ìƒì†ë˜ì–´ ì‚¬ìš©ê°€ëŠ¥í•˜ë‹¤.
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person): 
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute): # method overwriting: ë¶€ëª¨ classì—ì„œ ìƒì†í•œ methodë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  ì¬ì •ì˜í•´ì„œ ì‚¬ìš©.
        print('{}ì€ {}ë¶„ë™ì•ˆ ì¼ ë§ê³  ê³µë¶€í•œë‹¤.'.format(self.name, minute))

class Student(Person): # ë¶€ëª¨ class ê¸°ëŠ¥ë„ ì‚¬ìš©í•˜ê³  ì‹¶ì„ ê²½ìš°
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute): 
        print('{}ì€ {}ë¶„ë™ì•ˆ ì¼ ë§ê³  ê³µë¶€í•œë‹¤.'.format(self.name, minute)) # method overwriting
        super().work(minute)
     
bob = Student('Dasol', 20)
bob.eat('cherry')
bob.work(1)
```

# Class ì—°ì‚°ì ì¬ì •ì˜
- class ë˜í•œ ê¸°ë³¸ ì—°ì‚°ì²˜ëŸ¼ ì‚¬ìš©í•˜ê¸° ìœ„í•´ íŠ¹ë³„í•œ methodë¥¼ overwriting í•˜ëŠ” ê²ƒ
- print(ê°ì²´): 
    - __str__(self)
        - return '{}'.format(self.x)
    - __add__(self, point)
        - return Class(new_x, new_y)
    - __sub__(self, pt)
    - __mul__(self, factor)
        - return Point(self.x*factor, self.y*factor)
    - __len__(self)
    - __getitem__(self, index)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self): # print(ê°ì²´)ë¥¼ ê°€ëŠ¥í•˜ê²Œ í•˜ëŠ” ë©”ì„œë“œ
        return '({}, {})'.format(self.x, self.y)
    
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    
    def __mul__(self, factor):
        return Point(self.x*factor, self.y*factor)
    
    def __len__(self):
        return self.x+self.y
    
    def __getitem__(self, index):
        pass
    
p1 = Point(1,2)
print(p1)

p2 = Point(3,4)
# ADD
# p3 = p1.add(p2)
# print(p3)
P3 = p1 + p2
print(p3)

# Multiply
p4 = p1*3
print(p4)

# Length
print(len(p1))
```
