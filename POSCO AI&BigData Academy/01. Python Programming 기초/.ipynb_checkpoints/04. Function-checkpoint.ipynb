```python
def test(*args): # arguments
    for element in args:
        print(element)
        
test(1,2,3,4)
```

```python
def test(**kwargs): # key word arguments
    for key,value in kwargs.items():
        print(key, value)
        
test(a=1,b=2,c=3,d=4)
```
