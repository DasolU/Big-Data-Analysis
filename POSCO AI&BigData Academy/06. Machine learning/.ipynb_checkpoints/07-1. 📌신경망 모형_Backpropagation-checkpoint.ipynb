### 📌개념
- **error의 변화량에 weight가 어떤 영향을 주는가?**를 분석하기 위한 방법
  -  error = ouput layer의 반환 값 - target 값

### 📌과정
- error 변화량을 이전 layer로 역전파 시키며, 이 과정에서 weight의 영향을 받는다.

    - 📌중요한 점: **error의 변화량의 단위를 activation을 통과하기 전(v 단위)으로 변환해야한다**
    - error의 변화량의 단위는 activation을 통과한 x의 단위와 같다.
    - 그러나 weight의 단위는 activation을 통과하기 전인 v의 단위와 같다.
    - => error의 변화량의 단위를 activation을 통과하기 전으로 변환해야한다.
        - 아래 그림에서 **$e$**가 변환된 결과이다.
