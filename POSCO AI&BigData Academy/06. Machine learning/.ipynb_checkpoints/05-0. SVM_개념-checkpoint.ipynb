# SVM (Support Vector Machine)
### 필요성
데이터를 나눌 때, 데이터 분포가 정규분포가 아닌 경우 

### 방법
- margin을 최대화
  -  적당한 error를 허용하며 error 최소화
  
### 종류
- 종속 변수 형태에 따라 다르다.
- 범주형: 
    - support vector classifier
        - margin에 포함 or 반대 방향으로 분류된 점들 고려
- 연속형:
    - support vector regression 
        - margin 밖의 데이터에 대해서만 고려


# Lagrange multiplier
### 필요성
- 최적화 문제 풀 때
    - 목적 함수(f)를 최대화 하는 동시에 한정하고 싶은 조건(g)이 있는경우
    
### 개념
- f와 g가 접할 때, f의 극대/극소가 발생

- 극대/극소 지점: f와 g의 변화량이 상수배가 되는 지점 (변화하는 방향이 같을 때)
    - 따라서 아래식을 미분하여 0되는 지점이다.
    

# SVM 정의
- margin (M)을 최대화 = $\beta$ 최소화 + Error합 최소화
- 초평면(hyper plane): $f(x) = x^{T}\beta+\beta_{0}$
- decision rule
    - 초평면을 기준으로 y에 1 또는 -1을 부여하여 y와 평면의 곱을 양수로 만든다.
- Error를 허용


# SVM 계산




### lagrange multiplier 도입 이유
- lagrange function = cost function + decision rule(조건)
    - => lagrange function을 최소화 하기 위해 lagrange multiplier 도입
    
### 📌초평면의 기울기($\beta$)를 구하기
- 📌**초평면의 기울기($\beta$)를 구하기 위해 $\alpha$를 구한다.**
    - 방법: 위에서 구한 $\beta$, $\beta$, error를 lagrange function에 대입하고 이 식을 최소화하는 $\alpha$를 구한다.
    - KKT condition 가정: 

        - 이유: lagrange function을 최소화하는 $\alpha$가 Global minimum임을 보장
        
        - 의미: $\alpha$는 margin에 점이 위치할 때만 0이 아닌 값을 가질 수 있다.
            - Support vector: $\alpha$

# SVM vs. LDA 비교
### 장점
1. 데이터 분포 가정이 어려운 경우, covariance 구조 고려하는 것 비효율적
    - 데이터 분포 가정 없이 boundary 근처 관측치만을 고려
2. 예측 정확도가 높다.

### 단점
1. error term 앞 C를 결정해야함
    - C가 크면 error 허용 낮아진다.
2. 모형 구축 시간 오래 걸림
