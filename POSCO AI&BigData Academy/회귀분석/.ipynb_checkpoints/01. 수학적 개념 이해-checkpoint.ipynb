# 통계학이란
- 표본(sample)을 통해 모집단(Population)을 추정,추론 하고자 하는 학문
- 모수 (paramater): 수치로 표현되는 모집단의 특성
- 통계량 (Statistic): 표본 관측값에 의해 결정되는 양



# 확률
- 표본공간 (sample space): 모든 결과들의 모임
- 근원사건 (sample outcome): 표본 공간의 원소
- 사건 (event): 표본 공간의 부분집합. 근원사건의 집합
    - 배반 사건(mutually exclusive events): 서로의 교집합이 공집합인 사건
    
- 공분산
    - 두 변수 x,y의 선형적인 관계를 나타낸다.
    - 정의: E(x-m_x)(y-m_y)
        - 각 확률변수에 기댓값을 빼준 차이를 곱하여 평균을 취한다.
    - 의미:
        - 공분산이 양수: 
            - x와 y가 중심점으로부터 동일한 방향으로 움직인다.
        - 공분산이 음수: 
            - x와 y가 중심점으로부터 반대 방향으로 움직인다.
- 상관계수
    - 공분산을 단위화한 값 (-1<상관계수<1)
        - 공분산이 X,Y의 단위 크기에 영향을 받기 때문이다.


# 이산형 확률분포 (0 or 1을 나타내는 자료 분포)
- 베르누이 시행
    - 예시) 동전 던지기
- 이항분포
    - 성공확률이 p인 베르누이 시행을 독립적으로 n번 시행했을 때 성공한 횟수 분포
- 다항분포
    - 1회 시행 결과로 나올 수 있는 범주가 3개 이상 되는 확률 시험
- 포아송분포
    - 주어진 단위 구간 내에 평균적으로 발생하는 사건의 횟수가 정해져있을 때, 동일 단위에서 발생 횟수
    - 예시) 1시간 동안 걸려온 전화의 수

# 통계적 추론
- 점 추정 (Point estimation)
    - 추정량을 통해 모수를 추정
- 구간 추정 (intreval estimation)
    - 모수를 포함할 것으로 예상되는 구간을 제시
    - 신뢰 수준과 구간의 길이는 반비례


# 통계적 검정
- 대립 가설 (H1)
    - 주장하고자 하는 가설
- 귀무 가설 (H0)
    - 대립 가설의 반대 가설
    - 대립 가설의 입증에 사용
        - 귀무가설이 아니라는 증거를 데이터로부터 보임

# 오류
- 1종 오류
    - 귀무 가설이 맞을 때, 귀무 가설을 기각하는 오류
- 2종 오류
    - 귀무 가설이 틀렸을 때, 귀무 가설을 기각하지 않는 오류

# 검정 통계량, 기각역
- 검정 통계량
    - 표본에서 구해내는 값
        - 이 값을 기준으로 귀무 가설 기각 여부 결정
- 기각역 ($\alpha$ 값이 정해지면 정해지는 영역)
    - 검정 통계량의 분포 중 귀무 가설을 기각하는 구간
    - 기각역 결정 방법:
        - 1종 오류를 범할 가능성을 낮추기 위해 신중하게 선택해야 함.
- 검정 종류
    - 단측 검정: 목적에 따라 양 또는 음의 방향으로 검정
    - 양측 검정: 

# 유의 확률 (P-value)
- 📌대립가설을 더 선호하는 검정통계량 값이 나올 확률
    - 검정 통계량을 기준으로 귀무 가설에 반하는 모든 면적의 확률값을 더한 것
- 📌p-value 값이 $\alpha$ 값 보다 작다면=>귀무가설 기각, 대립가설 채택
    

# 귀무가설 기각 방법 정리
- 1.검정 통계량이 기각역보다 오른쪽에 있으면 기각
- 📌2.설정한 $\alpha$ 값 (유의수준)보다 p-value가 더 낮으면 기각

# 검정통계량과 관련된 분포
- Z 통계량
    - 귀무 가설이 "x의 평균이 m이다" 일 때, 귀무가설 하에서의 분포
    - 표본 정규분포: 중심이 0이고 분산이 1
- t 분포
    - 관측치의 수가 30개 이상으로 충분하다면 표본의 분산으로 대체 가능
    - 자유도가 클때만 정규분포에 근사
- 카이제곱 분포
    - 표준 정규 분포를 따르는 변수 Z의 제곱을 하면 자유도가 1인 카이제곱 분포를 따른다
    - 자유도가 커지면 skewness가 완화된다.
- F 분포
    - 서로 독립인 두 확률 변수의 정규분포를 자유도로 나누고 그것의 비율을 취한 것
    - 📌확률변수의 제곱합을 관측치로 나눈 것의 비율로 이루어진 통계량
        - 회귀분석에 응용

# likelihood function 

- Probability distribution function (확률분포함수; PDF): 
    - 정의: **모수를 알 때, 확률변수의 실현값을 예측하고자 함**
    - 예시) 모수(평균)이 주어졌을 때 x값이 나올 확률의 분포
    - 종류: 문맥에 따라 아래 3가지 의미로 사용되니 파악이 필요.
        - 1.확률밀도함수 (PDF; Probability density function)
            - 확률 변수가 연속형일 경우
        - 2.확률질량함수 (PMF; Probability mass function)
            - 확률 변수가 이산형일 경우
        - 3.누적분포함수 (Cumulative distribution function)
            - 누적 확률 분포함수

- 📌likelihood function (가능도함수): 
    - 정의: **확률변수의 실현값을 알 때(데이터가 있을 때), 모수를 측정하고자 함**
    - **모수($\mu$)가 변함에 따라 확률이 어떻게 변하는가를 살펴보고, likelihood를 최대로 만드는 모수의 값 (📌MLE; Maximum likelihoood Estimator)을 추정하는 것**
    - 예시) x들의 각각의 값이 주어졌을 때, 모수 $\mu$(평균)은 어떤 값일 가능성이 가장 높은가?
    - 확률분포함수와 동일한 함수이나 평균을 변수로 인식


# 📌MLE 
- Maximum likelihoood Estimator
    - 표현: 모수 위에 ^(hat)을 붙임
    - 구하는 방법: 미분
        - likelihood function가 $\mu$에 대한 함수이므로 likelihood function을 $\mu$으로 미분하여 0되는 값을 구한다

# 📌MLE 구하기
- log likelihood로 변환
    - 이유: exp 미분하기 용이함
    - log 함수는 단조증가하는 함수이기 때문에, log likelihood를 미분한 결과와 동일
