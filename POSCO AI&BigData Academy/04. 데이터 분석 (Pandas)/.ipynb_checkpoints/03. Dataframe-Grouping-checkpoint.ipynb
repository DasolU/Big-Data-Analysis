# group by
- column을 기준으로 통계적인 연산
- 데이터를 그룹화
    - SQL 의 group by 와 개념 동일
    - 그룹화 단계
        - 데이터 분할
        - operation 적용
        - 데이터 병합
- dataframe을 multi-level index로 바꾸기
    - df.set_index(['Pclass','Sex'])
    - multi-level index에서 row 가져오기
        - df.groupby(['Pclass','Sex']).mean().loc[(1,'female')]
- level과 함수 이용 방법
    - level
        - df.set_index('Age').groupby(level=0).mean()
    - 함수
        - df.set_index('Age').groupby(함수).mean()
    - aggregate (집계 함수)
        - df.set_index('Age').groupby(함수).aggregate([np.mean, np.sum, np.max])

```python
import numpy as np
import pandas as pd

# 숫자형 데이터를 범주형으로 변환
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10  

direc = '/Users/dasol/Documents/Job/포스코아카데미/강의자료/Part 01~04) Python/04. 데이터 분석을 위한 Python (Pandas)/data/train.csv'
df = pd.read_csv(direc)
df['Age'].apply(age_categorize)

direc = '/Users/dasol/Documents/Job/포스코아카데미/강의자료/Part 01~04) Python/04. 데이터 분석을 위한 Python (Pandas)/data/train.csv'
df = pd.read_csv(direc)

class_group = df.groupby('Pclass')
class_group
```

```python
# groups 
# key에 해당하는 index
class_group.groups # 속성
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

```python
class_group.count()
```

```python
class_group.mean()['Survived']
```

```python
# multi-level index
df.groupby(['Pclass','Sex']).mean()
```

```python
df.groupby(['Pclass','Sex']).mean().index
```

```python
# multi-level index에서 row 가져오기
df.groupby(['Pclass','Sex']).mean().loc[(1,'female')]
```

```python
# dataframe을 multi-level index로 바꾸기
df.set_index(['Pclass','Sex'])
```

## group by에서 level 이용 방법

```python
# group by에서 level과 함수 이용 방법
df.set_index('Age')
```

```python
# level=0은 첫 번째 index
df.set_index('Age').groupby(level=0).mean() 
```

```python
# multi-level index의 경우
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).mean() 
```

## group by에서 함수 이용 방법

```python
# 나이대 별 생존률 구하기
df.set_index('Age').groupby(age_categorize).mean() 
```

## 집계 함수 (aggregate)

```python
df.set_index('Age').groupby(age_categorize).aggregate([np.mean, np.sum, np.max])
```

## transform 함수
- 원본 dataframe의 index를 유지한채로 데이터 추가

```python
df.groupby('Pclass').transform(np.mean)
```

## 📌Pivot 함수
- dataframe의 형태 변경(reshaping)
- 예시) 
    - df.pivot('사용할 인덱스', '사용할 컬럼', 데이터)
        - 중복값이 있을 경우 ValueError를 반환
    - df.pivot_table(데이터,'사용할 인덱스', '사용할 컬럼', aggfunc =np.mean)
        - 중복 데이터를 처리해준다.
            - aggregate 옵션: aggfunc

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '월요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })
df
```

```python
df.pivot_table(df,'지역', '요일')
```

## 📌Stack & unstack 함수
- unstack: column index을 column 레벨로 올리는 것 (풀어주는 것)
- stack: row index를 column으로 내리는 것 (쌓는 것)

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })
new_df=df.set_index(['지역','요일'])
new_df
```

```python
# 지역 index를 column level로 올리기
new_df.unstack(0) 
```

```python
new_df.unstack(0).stack(0)
```

```python
# 요일 index를 column level로 올리기
new_df.unstack(1) 
```

```python
new_df.unstack(1).stack(0)
```

```python
new_df.unstack(1).stack(1)
```
