- pd.concat([df1, df2])
- index 유지
    - ignore_index=True 옵션: 0부터 연속적인 index를 가지게 함
- axis = 0 (위 아래) 또는 1 방향으로 병합 가능

## merge & join
- 특정 column을 기준으로 병합
- pandas.merge
    - pd.merge(df1, df2, on='column_Name')
    
- join
    - inner: 일치하는 것만 병합한다.
    - left: left에 있는 df1에 있는 데이터면 유지한다.
    - right:
    - outer: left와 right을 합친 것

```python
import numpy as np
import pandas as pd

customer = pd.DataFrame({'customer_id' : np.arange(6), 
                    'name' : ['철수'"", '영희', '길동', '영수', '수민', '동건'], 
                    '나이' : [40, 20, 21, 30, 31, 18]})


orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9], 
                    'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'], 
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
```

```python
customer
```

```python
orders.head()
```

```python
pd.merge(customer, orders, on='customer_id',how='inner')
```

```python
pd.merge(customer, orders, on='customer_id',how='right')
```

```python
# 가장 많이 팔린 아이템?
merged = pd.merge(customer, orders, on='customer_id',how='inner')
```

```python
merged.groupby('item').sum() 
```

```python
merged.groupby('item').sum()['quantity']
```

```python
merged.groupby('item').sum()['quantity'].max()
```

```python
merged['item'][merged.groupby('item').sum()['quantity'].max()]
```

```python
merged.groupby('item').sum().sort_values(by='quantity', ascending=False)
```
