# HTTP method
- HTTP 정의:
    - HTML 문서를 전송하는 규약(프로토콜)



- 클라이언트? 
    - 브라우저(크롬)
- HTTP 요청?
    - url 입력시 해당 서버에 요청됨
    
    - request method: Get, Post 등
- HTTP 응답?
    - 서버가 응답한 것을 클라이언트가 받아서 랜더링하여 사용자에게 제공
    
## Get 요청
- 데이터를 url에 포함하여 전달(리소스 요청)


## Post 요청
- 데이터를 form data에 포함하여 전달 (로그인에 사용)
