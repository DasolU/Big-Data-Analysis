# 단순 선형 회귀
- 정의
$y = \beta_{0}+\beta_{1}X + \epsilon$
    - coefficients (회귀계수)
        - $\beta_{0}$: intercept
        - $\beta_{1}$: slope
- 가정:
    - x와 y가 선형관계일 때
    - y 가 정규 분포일 때
    - noise가 정규 분포를 따를 때

# 선형 회귀의 정확도 평가
- MSE (mean squared error):
    - SSE를 표준화한 개념
- 변동성:
    - SST = SSE + SSR
        - SST: 총 변동 (y의 분산)
        - SSE: 회귀직선으로 설명 불가능한 변동
        - SSR: 회귀직선으로 설명 가능한 변동
        

- 📌$R^2$ 정확도 평가:
    - $R^2 = SSR \div SST$ = 총 변동 중에 회귀 직선으로 설명 가능한 변동 비율
    - 입력 변수 X로 설명할 수 있는 Y의 변동을 의미
    - 0초과 1이하 범위를 가지며, 1에 가까울수록 모형의 설명력이 높다는 의미
    - 쓸모없는 변수 개수가 많아질수록 $R^2$가 증가하기 때문에, $R^2$의 의미가 없어질 수 있다.

- 📌결국 중요한 점:
    - 정해진 Y의 변동성(분산)을 독립변수(X) 조합으로 얼마나 잘 설명하는가?
    - x로 Y를 설명하는 비율이 커질수록 $R^2$가 높아진다.
    

# 회귀계수에 대한 검정 
- 중요도 낮음, 해석 방법이 더 중요
- $\beta_{1}$ 검정
    - 귀무가설: $\beta_{1}=0$ (변수의 설명력이 0이다)
    - 대립가설: $\beta_{1}!=0$ (변수의 설명력이 존재한다 )
