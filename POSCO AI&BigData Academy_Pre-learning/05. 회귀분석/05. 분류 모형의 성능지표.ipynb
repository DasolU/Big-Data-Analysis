# 분류 모델에서 성능 지표

### confusion matrix (혼동행렬)


### 정확도 (Accuracy)
- 정의: 전체 데이터 중에 모형이 판단한 값이 실제 값과 부합하는 비율
- $\frac{옳게 분류된 데이터 수}{전체 데이터 수}$
    
### 정밀도 (Precision)
- $\frac{옳게 분류된 불량 데이터 수}{불량으로 예측한 데이터}$
- 분류 모형이 불량을 진단하기 위해 얼마나 잘 작동했는가

### 재현률 (Recall)
- $\frac{옳게 분류된 불량 데이터 수}{실제 불량 데이터 수}$
- 진단 확률: 불량 데이터 중 실제로 불량이라고 진단한 제품 비율

### 특이도 (Specificity)
- $\frac{옳게 분류된 정상 데이터 수}{실제 정상 데이터 수}$
- 분류 모형이 정상을 진단하기 위해 얼마나 잘 작동했는가
