# PCA 개념
- 차원 축소하며 정보의 손실을 최소화하는 방법
    - 방법: 더 적은 개수로 데이터의 변동을 가장 잘 설명하는 새로운 축 찾기
    - 예시) 새롭게 정의한 축으로부터 정사영 거리를 사용하여 데이터의 분산을 설명 가능해진다.
    
    
# PC 얻는 방법
- 공분산 구조 알아내기
    - = 공분산이 데이터 형태를 변형시키는 방향인 단축과 장축을 찾아내는 과정
- 가장 분산을 많이 설명하는 축: 공분산 타원의 장축
- 가장 분산을 적게 설명하는 축: 공분산 타원의 단축

# PC 의미?
- = PC score
- 새롭게 찾아낸 축에서의 좌표값
- = 새로운 축에 내린 정사영
