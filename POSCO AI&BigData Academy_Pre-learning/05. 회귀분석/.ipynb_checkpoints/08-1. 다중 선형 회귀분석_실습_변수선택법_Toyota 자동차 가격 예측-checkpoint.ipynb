```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
corolla = pd.read_csv('ToyotaCorolla.csv')
corolla.head()
```

```python
# 데이터 수와 변수의 수 확인하기
nCar = corolla.shape[0]
nVar = corolla.shape[1]
print(nCar, nVar)
corolla.shape
```

# 범주형 변수를 이진형 변수로 변환

```python
# 범주형 변수를 이진형 변수로 변환
corolla.Fuel_Type.unique()
## 가변수 생성
dummy_p = np.zeros(nCar)
dummy_d = np.zeros(nCar)
dummy_c = np.zeros(nCar)
```

```python
p_idx = np.array(corolla.Fuel_Type == 'Petrol')
d_idx = np.array(corolla.Fuel_Type == 'Diesel')
c_idx = np.array(corolla.Fuel_Type == 'CNG')
```

```python
p_idx
```

```python
dummy_p[p_idx] = 1
dummy_d[d_idx] = 1
dummy_c[c_idx] = 1
```

```python
Fuel = pd.DataFrame({'Petrol': dummy_p, 'Diesel': dummy_d, 'CNG': dummy_c})
Fuel
```

```python
# 불필요한 변수 제거 및 가변수 추가
corolla_ = corolla.drop(['Id', 'Model', 'Fuel_Type'], axis=1, inplace=False)
mir_data = pd.concat((corolla_, Fuel),1)
mir_data
```

# 설명 변수, 타겟 변수 분리 및 학습데이터와 평가데이터 분할

```python
# Target 변수를 제외한 변수 목록을 가져오기
feature_col = list(mir_data.columns.difference(['Price']))

x = mir_data[feature_col]
y = mir_data['Price']

train_x, test_x, train_y, test_y = train_test_split(x,y, train_size=0.7, test_size = 0.3) 
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
# OLS 적합/ 회귀모델 적합
# ordinary least square (최소 제곱 회귀)
full_model = sm.OLS(train_y, train_x)
fitted_full_model = full_model.fit()
# 결과 출력: summary 함수
# R-squared: 매우 높다.
# p-value: 대체로 낮다. => 대부분 변수들이 유의하다.
fitted_full_model.summary()
```

# VIF를 통한 다중공선성 확인

- p-value가 유의하지만 vif가 inf로 매우 높은 변수
    - 다중공선성이 높지만 유의하므로 제거하지 않는다.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF factor'] = [variance_inflation_factor(mir_data.values, i) for i in range(mir_data.shape[1])]
vif['features'] = mir_data.columns
vif
```

```python
# 학습 데이터의 잔차 확인
## Q-Q plot: 정규분포 확인
## 잔차가 얼마나 정규분포를 따르는가?
### 해석: 
resi = fitted_full_model.resid
fig = sm.qqplot(resi, fit=True, line='45')
```

```python
# residual pattern 확인
## predic value VS. residual
### 해석: 잔차들이 어느정도 "균등"하다.
###      잔차들이 y예측값에 따른 형태를 가지고 있지 않다.
pred_y = fitted_full_model.predict(train_x)
fig = plt.scatter(pred_y, resi, s=4)
plt.show()
```

```python
# 검증 데이터에 대한 예측
pred_y2 = fitted_full_model.predict(test_x)

# residual plot
plt.plot(np.array(test_y - pred_y2))
plt.show()
```
