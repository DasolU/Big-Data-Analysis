# SVD (Singular Value Decomposition)
## 정의
- matrix를 간단한 3개 요소로 나누는 것
 

### 기억할 점
- 1.양쪽 행렬을 제곱하면 identity matrix
- 2.가운데 행렬은 대각행렬이며 **대각 성분의 제곱**은 eigen value
- 3.오른쪽 행렬의 **column 성분**은 eigen vector

## 필요성    
- 계산을 속도를 높여준다.
- 📌SVD을 통해 matrix의 공분산 구조 행렬의 eigen value, eigen vector를 얻을 수 있다.
    - 이유: x가 centered 되어있다면 (평균이 0이라면), $x^{T}x$는 공분산 구조이기 때문.

## 특징
- 일반적인 nxp matrix의 decomposition이 가능하다. (matrix 형태에 대한 제약이 없다.)
