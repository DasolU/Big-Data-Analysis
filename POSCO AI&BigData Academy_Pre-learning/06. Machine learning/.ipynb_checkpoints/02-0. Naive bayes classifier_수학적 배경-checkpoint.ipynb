# Naive bayes classifier 수학적 배경
### 조건부 확률

### Bayes 정리
- 정의: 이전 경험과 현재 증거를 토대로 어떤 사건의 확률을 추론하는 과정



#### $P(A|B) = \frac{P(B|A)}{P(B)}$
- $P(B)$: B의 사전확률 (Prior, 과거의 경험)
- $P(B|A)$: 사건 A라는 증거에 대한 사후 확률 (Posterior, 사건 A가 일어났다는 것을 알 때, 그것이 사건 B로부터 일어난 것이라고 생각되는 조건부 확률)
- $P(A|B)$: 사건 B가 주어졌을 때, A의 조건부 확률 (likelihood, 관찰 결과에 기초한 A라는 가설에 대한 확률)

    - [추가 참고 문헌]: https://iludaslab.tistory.com/12

# Naive bayes classifier 개념



- 데이터(x)가 주어졌을 때 범주형 종속변수에 해당할 확률
    - (종속변수일 때, 데이터가 일어날 확률)  X (종속변수를 가질 확률 = 사전확률)
    - 이 확률을 가장 크게하는 C가 모델이 예측하는 범주가 된다.
    - 조건부 독립에 의해 (종속변수일 때, 데이터가 일어날 확률)이 쪼개져서 들어간다.
    - 분모는 동일하므로, 분자만을 비교한다.
        - 종속변수가 c가 일어날 확률 VS. 일어나지 않을 확률 중 확률이 큰 것을 선택한다. 
        


- 특징: 간단한 모델이지만 분류 정확도가 높다.
