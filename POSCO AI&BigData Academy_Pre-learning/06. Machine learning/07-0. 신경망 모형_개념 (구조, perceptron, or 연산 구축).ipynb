# 신경망 모형
- 인간의 뉴런의 정보 전달 과정을 모방
- 하나의 뉴런: 
    - 이전 뉴런에서 온 정보들을 계산하여 반환 =>  perceptron(인지)에 대응
- 시냅스: 
    - 뉴런과 뉴런의 연결 => layer에 대응

### 📌activation function 필요성:

# 바이너리 논리 연산
1. and 연산
    - 둘다 1일때만 1반환

2. or 연산
    - 둘 중에 하나라도 1이면 1반환

3. xor 연산 => perceptron 발전으로 이어진다.
    - 두 개가 일치할 때 0
    - 두 개가 불일치할 때 1

# Single-layer perceptron
- 속성이 2개일 때 perceptron: 직선
    - 직선을 기준으로 위/아래가 0/1으로 대응
- 직선으로 분류할 수 없는 경우 발생
    - 문제 해결 방법: 
        - 직선만을 사용하여 공간을 여러번 접어서 해결하는 방법 사용
        - => multi-layer perceptron (hidden layer) 도입

# OR 연산 구축
### weight 업데이터 알고리즘
- Gradient descent(경사하강법)
    - or 연산 결과 vs. perceptron 반환 값 비교
        - 일치할 때만 0
            - weight update X
        - 나머지는 1
            - 그리고 weight update 진행
        - weight update를 반복하여 모든 데이터를 잘 맞추면 update를 멈춘다.
