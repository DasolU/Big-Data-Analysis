# 딥러닝 발전 배경
## 신경망 모형의 한계점으로부터 시작

### Gradient vanishing
- sigmoid 함수의 미분함수의 최대값이 1/4이며, x값이 0에서 멀어질수록 0에 가까워진다.

- Backpropagation시 미분값이 매우 작아져 0으로 수렴
    - => error propagation시 x값이 크다면 error가 소멸되는 문제 발생
    
### 중간해 멈춤 현상
- local minimum에서 멈춰 global minimum찾지 못한다.
    - 초기값에 영향을 많이 받는다.
    
### 과적합 문제   

## 신경망 모형의 한계점 해결책
### ReLU (Rectified linear Unit)
- 0보다 큰 경우 기울기 일정
    - 즉, 미분값 유지
    - Gradient vanishing 해결
    - 중간해 멈춤 현상 완화

- 계산 간단
    - 학습 속도 매우 빠름

- 0보다 작은 경우
    - Leaky ReLU
        - 음수에서 0이 아닌 작은 값 부여
    - ELU (exponential linear unit)

### Pre-training: Boltzmann Machine
- 중간해 멈춤 현상 해결
- unsupervised 방식으로 미리 training시켜 local minima 문제 해결
- 올바른 초기값 선정
    - 방법: 인접한 layer 사이에서 x값을 예측, 예측한 값이 최소가 되는 weight를 찾아 초기값으로 설정
    

### Dropout
- 과적합 문제 완화
    - test error 상당히 감소 
- Hidden layer의 node를 임의의 확률에 따라 남김 (0.5~1사이 확률 권장)
- 계산 속도도 증가


### Weight decay
- 과적합 문제 완화
    - drop out 보다 전통적인 방법
- 방법: weight가 최대한 작은 값을 가지도록 penalty부여

# 은닉층 개수, 은닉 노드 개수
- test error 낮추는 방향으로 계속 테스트 해야한다.
    - 요즘은 auto ML이 많이 발전
- 참고 사항
    - 복잡한 문제-> 많은 layer
    - 다양한 입력 데이터-> 많은 node
    - 한 layer에 많은 node가 있는 경우 속도가 느려지며 효과가 적을 수 있다.
