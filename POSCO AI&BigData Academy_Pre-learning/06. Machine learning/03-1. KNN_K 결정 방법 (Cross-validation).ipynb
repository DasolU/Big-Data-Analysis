# K 결정 방법 
## Test error 를 작게하는 최적 k 선택


## 방법: Cross-validation 
### 개념
데이터 error를 과소추정하지 않으면서 데이터 손실을 최소화 하는 방법

### 필요성
trade-off 관계가 있는 두 가지를 해결할 수 있다.
1. overfitting
2. sample loss

### overfitting
- training error는 error를 과소추정한다.
    - training error가 낮은 모델을 선택하면 복잡한 모델을 선택하게 된다.
    - 따라서 test error를 구하고자 하면 문제가 발생한다.
        - 이유: 분리된 test 데이터를 만들면 train 데이터 loss 발생하기 때문
        
### 방법
- 데이터를 여러 개의 fold로 나눈뒤 test/ train set을 번갈아가며 사용하고 평균낸다.
- test error는 test fold에서 구한다.
